```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

```python
breast_cancer_data = sklearn.datasets.load_breast_cancer()
```

```python
#breast_cancer_data
```

```python
breast_cancer_dataframe = pd.DataFrame(breast_cancer_data.data, columns=breast_cancer_data.feature_names)
breast_cancer_dataframe.T
```

```python
breast_cancer_dataframe['label'] = breast_cancer_data.target
breast_cancer_dataframe.T
```

```python
breast_cancer_dataframe['label'].value_counts()

# Zero means malignant
# One means benign
```

```python
# Split the dataframe into X and Y
X = breast_cancer_dataframe.drop(columns='label', axis = 1)
Y = breast_cancer_dataframe['label'] 
```

```python
# Display the X feature columns
X
```

```python
# Display the target label
Y
```

```python
# Converting the feature columns and the target labels into numpy arrays
X = np.asarray(X)
Y = np.asarray(Y)
```

```python
X
```

# GridSearchCV is used to find the best parameters for our model

```python
model = SVC() # Loading the modelhttp://localhost:8888/notebooks/Hyperparameter%20Optimization.ipynb#GridSearchCV-is-used-to-find-the-best-parameters-for-our-model
```

```python
# Specify the model hyperparameters

hyperparameters = {
    'kernel': ['linear', 'poly', 'rbf', 'sigmoid'],
    'C': [1,5,10,20]
}
```

```python
classifier = GridSearchCV(model, hyperparameters, cv=5 )
```

```python
classifier.fit(X, Y)
```

```python
classifier.cv_results_
```

```python
# Display the best parameters
best_hyperparameters = classifier.best_params_
print(best_hyperparameters)
```

```python
# Display the results in a dataframe

results = pd.DataFrame(classifier.cv_results_)

results
```

```python
gridsearch_results = results[['param_C','param_kernel', 'mean_test_score', 'rank_test_score']]
gridsearch_results
```

# RandomizedSearchCV is also used to find the best parameters

```python
model = SVC() # Loading the model
hyperparameters = {
    'kernel': ['linear', 'poly', 'rbf', 'sigmoid'],
    'C': [1,5,10,20]
}
randomized_search_classifier = RandomizedSearchCV(model, hyperparameters, cv=5 )
randomized_search_classifier.fit(X, Y)
```

```python
randomized_search_classifier_results = pd.DataFrame(randomized_search_classifier.cv_results_)

randomized_search_classifier_results
```

```python
best_randomized_search_hyperparameters = randomized_search_classifier.best_params_
print(best_randomized_search_hyperparameters)
```

```python
randomized_search_results = randomized_search_classifier_results[['param_C','param_kernel', 'mean_test_score', 'rank_test_score']]
randomized_search_results
```
